# Graph Introduction

- Graph is a complex datastructure than trees
- A child node in a tree can only have one parent (or one possible way) but a graph's nodes can have multiple ways
- Graph is made of nodes and edges
  


- Another possible variant of the graph may include distances/cost of each node, thus called **Weighted Graph**. [eg. Locations in a map have distances associated to each edge]
  


- Another way, you may even add directions to this graph, thus called **Directed Graphs**



## Simple Implementation (using Dictionary)

```python
data = [
    ("Mumbai", "Paris"),
    ("Mumbai", "Dubai"),
    ("Paris", "Dubai"),
    ("Paris", "New York"),
    ("Dubai", "New York"),
    ("New York", "Toronto")
]
```

```python
# Imports
from typing import List, Dict
from pprint import pprint

class Graph:
    def __init__(self, nodes):
        self.nodes = nodes
        self.graph: Dict[str, List] = {}
        for key, value in nodes:
            if key in self.graph.keys():
                self.graph[key].append(value)
            else:
                self.graph[key] = [value]
        pprint(self.graph)
```

```python
graph = Graph(data)
```
